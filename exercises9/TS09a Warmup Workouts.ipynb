### Serial dependence

Now, suppose you want an experiment where the random variable $X_{t}$ is not serially independent.

Suppose it is auto-correlated and follows this first-order autoregressive, or AR(1), model:
$$
X_{t+1} = \rho X_{t} + \sigma\epsilon_{t+1}, \qquad \rho \in (0,1), \sigma \in [0,\infty),
$$
and $X_0$ is known. 

The exogenous shock is Gaussian: $\epsilon_{t+1} \overset{i.i.d.}{\sim} N(0,1)$ for all $t \geq 0$. Note that $\mathbb{E}_t(e_{t+1}) = 0$ and $\mathbb{E}_{t}(\epsilon_{t}) = \epsilon_{t}$ since $\epsilon_{t}$ is known at the start of date $t$---it is part of the information set against which the conditional expectations is form.

Here's a custom module we've baked for you earlier:

```python
def simulate_ar1(, , 
                 T=30, simulation="stochastic", 
                 DateShock=1, ShockUnit=1.0, y_initial=0.0):
    """Define AR(1) simulation method.
       
       Parameters
       ----------
       RHO: float64
           Persistence coefficient
       SIGMA: float64
           Std deviation of i.i.d. Gaussian shock 
       T: int
           Length of trajectory (including date t=0)
       simulation: str
           Default "stochastic"; "deterministic" 
           for deterministic impulse(s)
       DateShock: int (or list of ints)
           Specify timing of deterministic impulse
       ShockUnit: float64
           (Default is one s.d.: SIGMA). 
           Scale it as you like
       y_initial: float64
           Initial value of state, y_0
    """
    print("Message from SIMULATE_AR1:\n")
    if simulation == "stochastic":
        print("\t0. Stochastic simulation chosen: default is 1 s.d. shock")
         = np.random.randn(T)
    elif simulation == "deterministic":
        print("\t1. Deterministic Impulse Response chosen: default is 1 s.d. shock")
        print("\t2. If you want a 1 unit y-shock, set ShockUnit = 1.0/SIGMA\n")
         = np.zeros(T)
        [DateShock] = ShockUnit            
    y = np.zeros(T) # Pre-allocate memory space
    y[0] = y_initial
    for t in np.arange(T-1):
        y[t+1] = *y[t] + *[t+1]
    return y, 
```

The module ``simulate_ar1`` does the following:

* Requires two parameters describing the AR(1) model

* The rest of its inputs have default settings which you can change

* Depending on your ``simulation`` setting, the module generates either

    * a determinstic (``setting="deterministic"``) sequence of outcomes for a one-time increase in the level of the shock $\epsilon$ at a chosen date ``DateShock``

The AR(1) process exhibits a kind of limited *persistence of memory*---its past continues to influence its present realization but the long run effect of initial conditions disappear asymptotically (as time unfolds).

Notice how the blue-square sample path (the AR(1) process' sample realization) tends to swing longer in the same direction and takes longer to revert to its (zero) long-run mean?

In contrast, the red-circle sample path (the i.i.d. Gaussian process) is less persistent and mean-reverts quickly.

Also, the AR(1) process has more variance or is more noisy, as a consequence of the persistence parameter $\rho$. It can be [proven that its long-run or asymptotic variance is increasing with the parameters $\rho$ and $\sigma$](https://newonlinecourses.science.psu.edu/stat510/node/60/)---the theoretical, unconditional variance is $\text{var}(x_{t}) = \sigma^{2}/(1-\rho^{2})$.

## Problem 1 (Data and Model: NK interpretation of Inflation and Output Gap)

In this question we would like to get a handle on how a candidate macroeconomic theory can be mapped into empirically testable statistical or econometric models. We will start with a small example.



Let's assume that output gap $x$ (measured as percentage deviation of GDP from some potential path) is exogenously determined by the following deterministic process:
$$
x_{t+1} = \rho x_{t} + \sigma\epsilon_{t+1}, \qquad \rho \in (0,1), \sigma \in [0,\infty), \tag{1}
$$
where the exogenous shock is Gaussian: $\epsilon_{t+1} \overset{i.i.d.}{\sim} N(0,1)$ for all $t \geq 0$. 

Mathematically, this is a linear stochastic difference equation, also known as a first-order regressive model---or an AR(1) time-series process---to statisticians. If $\sigma=0$, then this is just a deterministic, autonomous linear difference equation.

In the 1990s economist began work on combining ideas of Keynes with New-classical macroeconomics (with its more detailed, microfounded stories of the macroeconomy). Marvin Goodfriend and Bob King called this the [New Neoclassical Synthesis](https://www.journals.uchicago.edu/doi/10.1086/654336). Later, this literature became known as the New Keynesian framework. This is now a workhorse for policy modelling and thinking at major institutions like the Board of Governors at the US Fed, the Bank of England, the Bank of Japan, the European Central Bank and many other institutions around the world. Similar models are also used in some specialized consultancies and investment banks that seek to predict and analyze government policy scenarios. (Although this model is not short in supply of logical and empirical deficiencies!)

Cutting a long story short, what is crucial in this framework is a modern version of the Phillips curve. In this model, firms are said to be imperfectly competitive and thus have market pricing power. However, they face some (exogenously specified cost or friction to changing prices). As a result, in an aggregate equilibrium, one can derive a characterization that looks like the Old Keynesian Phillips curve. A key difference here is that inflation today may sensitively depend on *expectations of inflation* into the future. (Hence the language of most inflation-targeting central banks today that often discuss the need to "anchor inflation expectations" in their policy stance.) Here's a [readable exposition on this model](https://www.richmondfed.org/-/media/richmondfedorg/publications/research/economic_quarterly/2008/fall/pdf/hornstein.pdf).

After some math, a simple version of the New Keynesian Phillips curve looks like this:
$$
\pi_{t} = \beta \mathbb{E}_{t}[\pi_{t+1}] + \kappa x_{t}, \tag{2}
$$
where $\beta \in (0,1)$ represents agents' subjective discount factor, and, $\kappa > 0$ is a composite parameter which further depends on micro-parameters on taste and production technology. The $\mathbb{E}_{t}[\cdot]$ notation refers to a conditional expectations operator, where conditioning is with respect to the available date-$t$ information set in the model (note: $x_{t}$ is known/realized at the beginning of date $t$).

### Answer

You should have a graph that is affine and upward sloping in $(x_{t}, \pi_{t})$-space.

Its vertical intercept should be the value of $\beta \mathbb{E}_t\pi_{t+1}$.

Instead of sketching it, let me plot a parametric instance of this. First, let's parametrize and set up the PC model using ``Python``'s ``lambda`` (anonymous function):

```python
# Generate evenly-spaced grid for sample x_t points
x_grid = np.linspace(-1.0, 1.0, 30)

# Parameters
 = 0.99
 = 0.086

# Define PC model "NKPC"
# Note: 'lambda' allows us to define a function on the go
nkpc = lambda x, epinext: *epinext + *x
```

Here the ``lambda`` anonymous function method has created a function called ``nkpc``. It takes in two variable inputs: ``epinext`` (for expected next-period inflation) and ``x`` (for current output gap).

We can evaluate this function, for example:

```python
# Evaluate nkpc at particular (x, E_pi_+1)
E_ = 0.01
x = 0.5
 = nkpc(x, E_)

# Display current inflation according to PC model
print()
```

Let's define a function to plot the graph of this ``nkpc`` model:

```python
def nkpc_plot(E_):

    # label axes
    plt.xlabel("Output gap, $x_{t}$")
    plt.ylabel("Current inflation, $\pi_{t}$")

    # NKPC graph - evaluate
    pi_graph = nkpc(x_grid, E_)

    # NKPC graph - plot
    plt.plot(x_grid, pi_graph,
             label = "NKPC at $\mathbb{E}_{t}\pi_{t+1} =$" + str(E_))

    # Reference axes (horizontal and vertical dash-dots)
    plt.plot(x_grid, 
             np.zeros(x_grid.size), 
             '-.k', alpha=0.5)
    plt.plot(np.zeros(10), 
             np.linspace(pi_graph.min(), pi_graph.max(), 10), 
             '-.k', alpha=0.5)

    # Render a legend box based on label(s)
    plt.legend()
```

Now let's plot it:

```python
# Invoke plotting function
nkpc_plot(E_)

# output to screen
plt.show()
```

From the interactive experiment above you can see:

* Higher expected future inflation will increase current inflation at every given level of output gap today.

* Lower expected future inflation will decrease current inflation at every given level of output gap today.

The intuition for this in the NKPC model is that:

* Businesses or firms are forward looking - they maximize total expected discounted profits

* Future profits depend on both future demand and production cost conditions. These, in turn, depend on future inflation.

* All else constant, if the economy expects higher inflation next period, firms will want to raise prices today and thus inflation today jumps in response. (But by how much?)

* However, in this model, in the short run, adjusting prices immediately today is costly. That is, suppose inflation has been $1\%$ (per annum) for a while. If expected inflation remains at $1\%$, firms may not all adjust prices such that inflation today hits $\beta \times 1\%$, but instead firms may want to adjust prices today by less in response, say by $-\kappa\%$, in response to slacker demand conditions today (i.e., current output gap is at $-1\%$). 

### Question 1.2

Since the NKPC is forward looking (i.e., $\pi_{t}$ depends on a projection of its future value conditional on available current information) and it also depends on an non-autonomous terms $x_{t}$, $\pi_{t}$ has no initial value. (This is akin to our earlier exercise on the RBC model where consumption had no initial condition and is a jump variable. Also since $1/\beta > 1$, the backward solution will be unstable and explosive.)

Start with a date $t$ and solve this equation forward up to some arbitrary future horizon $t+k > t$. If you're rusty on solving difference equations, [catch up here!](#revision-self) (esp. Homework 6 on forward difference equations).

Show that the dynamic equilibrium of this model is given by the two equations:

$$
x_{t+1} = \rho x_{t} + \sigma\epsilon_{t+1}, \qquad \rho \in (0,1), \sigma \in [0,\infty),
\tag{1}
$$

where $\epsilon_{t+1} \overset{i.i.d.}{\sim} N(0,1)$ for all $t \geq 0$, and,

$$
\pi_{t} = \frac{\kappa}{1-\beta\rho}x_{t} \tag{2'},
$$
for some given initial state $x_0$.

*Hint*: The only source of randomness in this example comes from the stochastic process of $\{x_{t}\}$. Inflation would inherit from this randomness. From the AR(1) model representing $x_{t}$, we can apply the law of iterated expectations: $\mathbb{E}_{t}\left[ \mathbb{E}_{t+1} \left[ \cdots \mathbb{E}_{t+k-1}x_{t+k}\right]\right] = \mathbb{E}_{t}x_{t+k}$ for any $k \geq 2$. Likewise, $\mathbb{E}_{t}\left[ \mathbb{E}_{t+1} \left[ \cdots \mathbb{E}_{t+k-1}\pi_{t+k}\right]\right] = \mathbb{E}_{t}\pi_{t+k}$ for any $k \geq 2$.

What is the difference between the system (1) and (2') with the system (1) and (2)?

### Answer

Now we solve for the (rational expectations) equilibrium of this simple model.

Since $\beta > 1$, iterative backward substitution of $\pi_{t}$ will imply an explosive solution (the inflation process will wander off to $\pm \infty$. So this cannot be a well-defined equilibrium.

We solve it forward. You can show that condition (2) can be iteratively substituted forward to get:

$$
\pi_{t} = \beta^{k} \mathbb{E}_{t}\pi_{t+k} + \kappa \sum_{j=0}^{k}\beta^{j}\mathbb{E}_{t}x_{t+j}.
$$

(Do the algebra and then apply the law of iterated expectations.)

This says that inflation today depends on the forecast of itself into the future, and also on some distributed lag of forecasts of output gap (i.e., demand conditions) into the future.

Consider the limit of $k \rightarrow \infty$ (the degree of "forward-looking-ness" is infinitely far ahead):

$$
\lim_{k \rightarrow \infty} \pi_{t} = \lim_{k \rightarrow \infty} \left\{\beta^{k} \mathbb{E}_{t}\pi_{t+k} + \kappa \sum_{j=0}^{k}\beta^{j}\mathbb{E}_{t}x_{t+j}\right\}.
$$

Since $\beta < 1$, the last equation becomes

$$
\begin{align}
\pi_{t} &= \kappa \sum_{j=0}^{\infty}\beta^{j}\mathbb{E}_{t}x_{t+j}
\\
&= \kappa  \sum_{j=0}^{\infty}\beta^{j}\left[ \rho^{j}x_{t} + \sigma \sum_{i=0}^{j}\rho^{i}\mathbb{E}_{t}\epsilon_{t+i+1}\right].
\end{align}
$$
The last line is a result from using the backward substitution solution for the AR(1) process of $x_{t}$.

Since $\rho < 1$, and, from the stated fact that $\mathbb{E}_{t}\epsilon_{t+j} = 0$ for all $j\geq 1$, we can simplify the last line to obtain (2'). Note: the first terms can be simplified by noting that at date $t$, $\mathbb{E}_{t}x_{t} = x_{t}$ since output gap is already realized; then the first term just an infinite geometric series, since $0< (\beta \rho)^{j} < 1$, for all $j \geq 0$.

Discussion:

* This solution encodes the following stringent ontological concept: If agents (and firms) in the model are forward-looking in a *rational-expectation* sense---i.e., they understand and incorporate the equilibrium's underlying probability laws, which in this case is the process for $x_{t}$---then the *rational expectations equilibrium* solution is self-enforcing in terms of actions and beliefs.

* (Of course, the epistemological question of how people come to learn or precisely understand the probability laws of the environment they operate and make decisions in is a mysterious one in the rational expectations paradigm. The rational expectations concept, akin to the concept of Walrasian equilibrium in general equilibrium theory, is a tautological one: Beliefs are already in a fixed point with equilibrium, objective, or model-consistent probabilities.)

Remark:

* Conditions (1) and (2') constitute the (rational expectations) recursive competitive equilibrium solution of the model. This system in turn can be solved or evaluated to obtain the equilibrium outcomes.

* Conditions (1) and (2) form the characterization of equilibrium. That is, a (rational expectations) recursive competitive equilibrium, in this model, must satisfy these two equations.

Let's answer this in terms of an output-gap and inflation Phillips curve metaphysical construct. The observation by the BIS paper (see Claudio Borio's speech), from the lens of our setup here, is that the "empirical" Phillips curve has been flattening out. In terms of our highly-stylized theory here, this could only mean that the slope term in solution (2') is flattening out to zero. But how, if the parameters are unchanged?

Short of just changing parameters arbitrarily, here's a possible conjecture of what may be going in the BIS evidence vis-a-vis our model:

* The model, equations (1) and (2) above, only considers the forward looking behavior of firms, in terms of pricing their goods. The demand side was treated as an exogenous process.

    * One may not be able to detect an upward-sloping empirical PC using empirical regressions

    * Successful monetary policy could have accounted for the coexistence of these two things: an empirical observation and a metaphysical construct.
    
* However, that is not the last word on the PC. Other economists would argue that there may be no logical foundation for the conceptual (structural) New Keynesian PC. After all, the existence of this NKPC relies on strong and exogenous assumption that firms cannot change prices quickly---an ad hoc behavioral assumption. For example, [some "New Monetarist" economists](https://doi.org/10.1111/j.1542-4774.2012.01081.x) have shown that one could write down a theory where the observed heterogeneity in how slow or quickly firms adjust prices in the real-world micro-data is a consequence of firms with complete freedom to reset prices any time; except that in equilibrium they may not find it optimal to do so all the time. However, in their model, monetary policy is neutral: There is no inflation and output gap Phillips curve at all: neither structurally nor empirically.

* This is logically troubling, as far as central bank monetary policy done from a Keynesian perspective is concerned. Maybe that is why central bank economists like Jim Bullard still prefer to work within the logical paradigm of an NKPC.

```python
def lde(y, a, b):
    """First-order linear difference equation
    a: coefficient on lag y
    b: constant term
    """
    ynext = a*y + b
    return ynext

def lde_steadystate(a, b):
    """First-order linear difference equation
    a: coefficient on lag y
    b: constant term
    """
    ystar = b/(1.0 - a)
    return ystar

# Loop to generate trajectory given model and parameters:
def lde_trajectory(a, b, y0, T):
    """Simulate the outcome path of the LDE"""
    y = y0
    path = [ y ]
    for t in range(T):
        # Evaluate LDE at current y := y(t)
        ynext = lde(y, a, b)
        # Store outcome in path
        path.append(ynext)
        # Update y(t) for next period (iteration)
        y = ynext  
    return path
```

```python
# First example
a = 0.3
b = 1.0

# Steady state
ystar = lde_steadystate(a,b)
print('Steady-state y* =%6.3f' %(ystar))
```

```python
# Simulate T_sim+1 length path conditional on initial value of y0 < ystar
y0 = .5*ystar
T_sim = 10
path = lde_trajectory(a, b, y0, T_sim)
```

```python
def plot_phasediagram(path, ystar, a, b):
    """Module to plot phase diagram and superimpose sample path"""
    # Sample domain and range for LDE graph
    X = np.linspace(min(path)*0.99, 1.10*max(path), 20)
    Y = lde(X, a, b)

    plt.figure()
    # Axes labels
    plt.xlabel("$y_{t}$")
    plt.ylabel("$y_{t+1}$")
    # Plot LDE graph
    plt.plot(X,Y, '-b', label="Model")
    plt.plot(X,X, '-', color='gray', label='$45^{\circ}$ line')
    # Plot trajectory in same diagram
    plt.plot(ystar,ystar, 'og', label="Steady state")
    for t in range(len(path)-1):
        x, y = path[t], path[t+1]
        plt.plot(x,y,'s')
        plt.text(x,y,"%i" % (t))
    plt.legend()
```

```python
plot_phasediagram(path, ystar, a, b)
```

```python
def plot_trajectory(path, ystar):
    """Module to plot time path"""
    # Visualize results
    plt.figure()
    # Steady-state path
    plt.plot(np.tile(ystar,len(path)), '-r')
    # Plot path of LDE outcome from y0
    for t in range(len(path)-1):
        plt.plot(t, path[t], 's')
    plt.xlabel('Time, $t$')
    plt.ylabel('$y_{t}$')
```
